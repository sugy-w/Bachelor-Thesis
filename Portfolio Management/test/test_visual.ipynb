```python
import sys
import os

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
naive = []
with open("naive.txt", "r") as file:
    content = file.read()
    for i in content.split():
        naive.append(float(i))
```

```python
arima = []
with open("arima.txt", "r") as file:
    content = file.read()
    for i in content.split():
        arima.append(float(i))
```

```python
lasso = []
with open("lasso.txt", "r") as file:
    content = file.read()
    for i in content.split():
        lasso.append(float(i))
```

```python
nasdaq = []
with open("nasdaq.txt", "r") as file:
    content = file.read()
    for i in content.split():
        nasdaq.append(float(i))
```

```python
ridge = []
with open("ridge.txt", "r") as file:
    content = file.read()
    for i in content.split():
        ridge.append(float(i))
```

```python
snp = []
with open("snp500.txt", "r") as file:
    content = file.read()
    for i in content.split():
        snp.append(float(i))
```

```python
dates = []
with open("dates.txt", "r") as file:
    content = file.read()
    for i in content.split("\n"):
        dates.append(pd.to_datetime(i))
```

```python
plt.plot(dates[35:], nasdaq, linestyle="dotted", label="NASDAQ")
plt.plot(dates[35:], snp, linestyle="dotted", label="S&P 500")

plt.plot(dates[35:], naive, label="NAÏVE")
plt.plot(dates[35:], arima, label="ARIMA")
plt.plot(dates[35:], ridge, label="Ridge")
plt.plot(dates[35:], lasso, label="Lasso")

plt.ylabel("Portfolio Values ($)")

plt.xticks(rotation=45)

plt.legend()
```
